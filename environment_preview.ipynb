```python
import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt
import minigrid
import time
from gymnasium.envs.registration import register

```

```python
from gymnasium.envs.registration import register

register(
    id="MiniGrid-DoorKey-5x5-Enemy-v0",
    entry_point="enemy_doorkey_env:DoorKeyWithEnemyEnv",
    kwargs={"size": 6}
)


env = gym.make("MiniGrid-DoorKey-5x5-Enemy-v0", render_mode="rgb_array")
print(env.unwrapped.width, env.unwrapped.height)
# or:
print(env.get_wrapper_attr("width"), env.get_wrapper_attr("height"))

def show(frame):
    plt.figure(figsize=(4,4))
    plt.axis("off")
    plt.imshow(frame)
    plt.show()
```

```python
obs, info = env.reset()
frame = env.render()
show(frame)

for t in range(20):
    action = env.action_space.sample()
    obs, reward, terminated, truncated, info = env.step(action)
    frame = env.render()
    show(frame)
    
    if terminated or truncated:
        obs, info = env.reset()
```

```python
env.close()
```
